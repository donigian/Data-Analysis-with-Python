```python
all_users = users.append(answerers,ignore_index=True)
HTML(all_users.to_html())
```

```python
# boolean series of all duplicates
all_users.duplicated()
```

```python
# remove dups, drop_duplicates()
# all_users = all_users[-all_users.duplicated()]
all_users.drop_duplicates()
```

```python
HTML(posts.head(1).to_html())
```

```python
HTML(all_users.head(1).to_html())
```

```python
# join posts data with all_users, left => posts, right=> all_users
# overlapping column names are appended with _x, _y
all_users_posts = pd.merge(posts, all_users, left_on = 'OwnerUserId', right_on='Id')
all_users_posts
HTML(all_users_posts.head(1).to_html())
```

```python
# merge appends a suffix _x, _y for duplicate columns
# reference http://pandas.pydata.org/pandas-docs/dev/merging.html
uposts = pd.merge(posts, all_users, left_on = 'OwnerUserId', right_on='Id', suffixes=['_post', '_user'])
uposts
HTML(uposts.head(1).to_html())
```

```python
uanswers = pd.merge(answers, all_users, left_on = 'OwnerUserId', right_on='Id', suffixes=['_post', '_user'])
uanswers
```

```python
# pandas objects can be split on any of their axes. 
# The abstract definition of grouping is to provide a mapping of labels to group names. 
# reference http://pandas.pydata.org/pandas-docs/dev/groupby.html
uanswers.groupby('DisplayName').Score.mean().order()
```

```python
# Once the GroupBy object has been created, several methods are available to perform a computation on the grouped data. 
mean_count = uanswers.groupby('DisplayName').Score.agg(['mean','count','min','max','std']) 

HTML(mean_count.sort('count').tail(10).to_html())
```
